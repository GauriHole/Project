```python
import requests
from bs4 import BeautifulSoup

url = "https://www.meesho.com/a1-dresses/p/6hnkb2"

# Define headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Make the request with headers
page = requests.get(url, headers=headers)

# Parse the page content
soup = BeautifulSoup(page.content, "html.parser")

# Pretty print the soup
print(soup.prettify())
```

```python

review=soup.find_all('div',class_="sc-ftTHYK eHVGcU")
review
rating=soup.find_all('div',class_="sc-ftTHYK eHVGcU")
rating
```

```python
review1=soup.find_all('span',class_="sc-eDvSVe gUjMRV Comment__CommentText-sc-1ju5q0e-3 cfdxfJ Comment__CommentText-sc-1ju5q0e-3 cfdxfJ")
review1
```

```python

rating1=soup.find_all('span',class_="sc-eDvSVe laVOtN")
rating1
```

```python

review_text=[review.text.strip() for review in review1]
review_text
```

```python
ratingcnt = [rate.text.strip() for rate in rating1]
print(ratingcnt)
```

```python

for review in review1:
    print(review.get_text(strip=True))
    
```

```python

for rate in rating1:
    print(rate.get_text(strip=True))
    
```

```python
print(f"Number of reviews: {len(review_text)}")
print(f"Number of ratings: {len(ratingcnt)}")
```

```python
print("Reviews: ", review_text)
print("Ratings: ", ratingcnt)
```

```python
import pandas as pd
# Truncate the longer list (ratings) to match the length of the shorter list (reviews)
min_length = min(len(review_text), len(ratingcnt))

# Truncate both lists
review_text = review_text[:min_length]
ratingcnt = ratingcnt[:min_length]

# Create the DataFrame
df = pd.DataFrame({
    'Rating': ratingcnt,
    'Review': review_text
})

# Print the DataFrame
print(df)

# Save the DataFrame to CSV
df.to_csv('reviews_and_ratings.csv', index=False)
print("DataFrame saved as 'reviews_and_ratings.csv'.")
```
